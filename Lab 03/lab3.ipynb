## Zadanie 1 - odległość edycyjna

```python
def edit_distance(s1, s2):
    """Compute the edit distance between two strings."""
    n = len(s1)
    m = len(s2)
    # Initialize the matrix.
    D = [[0 for j in range(m + 1)] for i in range(n + 1)]
    # Fill in the base cases.
    for i in range(n + 1):
        D[i][0] = i
    for j in range(m + 1):
        D[0][j] = j
    # Fill in the rest of the matrix.
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if s1[i - 1] == s2[j - 1]:
                D[i][j] = D[i - 1][j - 1]
            else:
                D[i][j] = min(D[i - 1][j - 1], D[i - 1][j], D[i][j - 1]) + 1
    return D[n][m]

def get_edit_path(s1, s2):
    """Compute the edit path between two strings."""
    n = len(s1)
    m = len(s2)
    # Initialize the matrix.
    D = [[0 for j in range(m + 1)] for i in range(n + 1)]
    # Fill in the base cases.
    for i in range(n + 1):
        D[i][0] = i
    for j in range(m + 1):
        D[0][j] = j
    # Fill in the rest of the matrix.
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if s1[i - 1] == s2[j - 1]:
                D[i][j] = D[i - 1][j - 1]
            else:
                D[i][j] = min(D[i - 1][j - 1], D[i - 1][j], D[i][j - 1]) + 1
    # Initialize the path.
    path = []
    # Fill in the path.
    i = n
    j = m
    while i > 0 or j > 0:
        if i > 0 and j > 0 and D[i][j] == D[i - 1][j - 1] and s1[i - 1] == s2[j - 1]:
            path.append(('M', s1[i - 1]))
            i -= 1
            j -= 1
        elif i > 0 and D[i][j] == D[i - 1][j] + 1:
            path.append(('D', s1[i - 1]))
            i -= 1
        elif j > 0 and D[i][j] == D[i][j - 1] + 1:
            path.append(('I', s2[j - 1]))
            j -= 1
        else:
            path.append(('R', s2[j - 1]))
            i -= 1
            j -= 1

    return path[::-1]

def print_edit_path(s1, s2):
    """Print the edit path between two strings."""
    path = get_edit_path(s1, s2)
    new_str = s1[:]
    i = 0
    for op, c in path:
        if op == 'M':
            new_str = new_str[:i] + c + new_str[i + 1:]
        elif op == 'D':
            print(f"{new_str} -> {new_str[:i] + new_str[i + 1:]} (delete {new_str[i]} at index {i})")
            new_str = new_str[:i] + new_str[i + 1:]
            i -= 1
        elif op == 'I':
            print(f"{new_str} -> {new_str[:i] + c + new_str[i:]} (insert {c} at index {i})")
            new_str = new_str[:i] + c + new_str[i:]
        elif op == 'R':
            print(f"{new_str} -> {new_str[:i] + c + new_str[i + 1:]} (replace {new_str[i]} with {c} at index {i})")
            new_str = new_str[:i] + c + new_str[i + 1:]

        i += 1
```

## Zadanie 2 - najdłuższy wspólny podciąg

```python
from spacy.tokenizer import Tokenizer
from spacy.lang.pl import Polish
nlp = Polish()
# Create a blank Tokenizer with just the Polish vocab
tokenizer = Tokenizer(nlp.vocab)
```

```python
def longest_common_subsequence(s1, s2, get_common_subsequence=False):
    """Compute the length of the longest common subsequence of two strings."""
    n = len(s1)
    m = len(s2)
    # Initialize the matrix.
    D = [[0 for j in range(m + 1)] for i in range(n + 1)]
    # Fill in the rest of the matrix.
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if s1[i - 1] == s2[j - 1]:
                D[i][j] = D[i - 1][j - 1] + 1
            else:
                D[i][j] = max(D[i - 1][j], D[i][j - 1])

    if get_common_subsequence:
        # Initialize the path.
        path = []
        # Fill in the path.
        i = n
        j = m
        while i > 0 and j > 0:
            if s1[i - 1] == s2[j - 1]:
                path.append((s1[i - 1], i-1, j-1))
                i -= 1
                j -= 1
            elif D[i - 1][j] > D[i][j - 1]:
                i -= 1
            else:
                j -= 1

        return D[n][m], path[::-1]

    return D[n][m], None
```

```python
with open('romeo-i-julia-700.txt') as f:
    text = f.read()

tokens = [token for token in tokenizer(text)]
```

```python
def get_reduced_tokens(tokens, percentage=0.97):
    for token in tokens:
        if np.random.random() < percentage:
            yield token
```

```python
token1 = get_reduced_tokens(tokens)
token2 = get_reduced_tokens(tokens)

with open('text1.txt', 'w', encoding='utf-8') as f:
    for token in token1:
        f.write(token.text_with_ws)

with open('text2.txt', 'w', encoding='utf-8') as f:
    for token in token2:
        f.write(token.text_with_ws)
```

```python
def diff_tool(text1_filename ,text2_filename):
    with open(text1_filename) as f:
        text1 = f.readlines()

    with open(text2_filename) as f:
        text2 = f.readlines()

    text1_index = text2_index = 0

    while text1_index < len(text1) and text2_index < len(text2):
        text1[text1_index] = text1[text1_index].strip()
        text2[text2_index] = text2[text2_index].strip()

        lcs, cs = longest_common_subsequence(text1[text1_index], text2[text2_index], True)

        if lcs == len(text1[text1_index]) == len(text2[text2_index]):
            text1_index += 1
            text2_index += 1
            continue
        
        if len(cs) <= 0:
            print(repr(f"[{text1_index + 1}] < {text1[text1_index]}"))
            print(repr(f"[{text2_index + 1}] < {text2[text2_index]}"))
            text1_index += 1
            text2_index += 1
            continue

        start_text1 = cs[0][1]
        end_text1 = cs[-1][1]
        start_text2 = cs[0][2]
        end_text2 = cs[-1][2]

        different_text1 = f"{text1[text1_index][:start_text1]}{text1[text1_index][end_text1+1:]}".strip()
        # print(repr(f"[{text1_index + 1}] < {different_text1}"))

        print(repr(f"[{text1_index + 1}] < {text1[text1_index]}"))

        different_text2 = f"{text2[text2_index][:start_text2]}{text2[text2_index][end_text2+1:]}".strip()
        # print(repr(f"[{text2_index + 1}] > {different_text2}"))

        print(repr(f"[{text1_index + 1}] < {text2[text2_index]}"))

        text1_index += 1
        text2_index += 1

    while text1_index < len(text1):
        print(repr(f"[{text1_index + 1}] < {text1[text1_index].strip()}"))
        print(f"[{text1_index + 1}] > EOF")
        text1_index += 1

    while text2_index < len(text2):
        print(f"[{text2_index + 1}] < EOF")
        print(repr(f"[{text2_index + 1}] > {text2[text2_index].strip()}"))
        text2_index += 1
```
